Importing The Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#Loading Dataset into Pandas Dataframe
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
credit_card_data.shape
```

```python
#First and Last 5 rows of the dataset
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
#Dataset Information
credit_card_data.info()
```

```python
#Checking the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
#distribution of legit and fraud transaction
credit_card_data['Class'].value_counts()
```

This Dataset is highly unbalanced

0 - Legit Transaction


1 - Fraud Transaction

```python
#Separating the data for analysis
legit = credit_card_data[credit_card_data.Class==0]
fraud = credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both the transactions
credit_card_data.groupby('Class').mean()
```

Under-Sampling


Build a sample dataset containing similar distribution of legit and fraud transactions


```python
legit_sample = legit.sample(n=52)
```

Concatenating two dataframes

```python
new_dataset = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.shape
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

Now the Dataset is balanced

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into Features & Targets

```python
X = new_dataset.drop(columns='Class',axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split the data into Training Data & Testing Data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

X_train = contains all the features of the training data

Y_train = contains the corresponding labels

X_test = contains all the features of the testing data

Y_test = contains the corresponding labels


```python
#after splitting the data
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression for Binary Classfication

```python
model =  LogisticRegression()
```

```python
#training the Logistic Regression model with Training Data
model.fit(X_train, Y_train)
```

Model Evaluation

-based on accuracy score

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on Training Data is: ',training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy on Test Data is: ',test_data_accuracy)
```
